## **문제 06 실패율**

https://school.programmers.co.kr/learn/courses/30/lessons/42889

```python
## 정답 코드

def solution(N, stages):
    answer = []
    people = len(stages)
    for i in range(1, N+1):
        if people == 0:
            answer.append([i, 0])
        else:
            per = stages.count(i)/people
            answer.append([i, per])
            people -= stages.count(i)

    answer.sort(reverse = True, key = lambda x:x[1])
    answer = [ans[0] for ans in answer]
    return answer
```

```python
n = 5
stages = 	[2, 1, 2, 6, 2, 4, 3, 3]

solution(n, stages)
```

```python
n = 4
stages = 	[4,4,4,4,4]

solution(n, stages)
```

```python
## 풀리는 코드이긴 한데 자꾸 시간초과가 뜬다,.... 일부러 시간복잡도 높이기 싫어서 pop 안쓰려고 한건데

def solution(s):
    answer = -1
    s_lst = [i for i in s]
    cnt = 0
    while cnt <= len(s):
        if len(s_lst) == 0:
            answer = 1
            break
        for idx in range(1, len(s_lst)):
            if s_lst[idx] == s_lst[idx-1]:
                s_lst[idx], s_lst[idx-1] = 0, 0
        s_lst = [s for s in s_lst if s != 0]
        cnt += 1
    if answer == -1:
        answer = 0
    return answer

s = 'cdcd'
solution(s)
```

```python
## 풀리는 코드이긴 한데 자꾸 시간초과가 뜬다,.... 일부러 시간복잡도 높이기 싫어서 pop 안쓰려고 한건데


##while문이랑 deque 써서 popleft 진행하자
from collections import deque
def solution(s):
    answer = -1
    s_lst = [deque(i for i in s)]
    cnt = 0
    while cnt <= len(s):
        lst = [deque()]
        if len(s_lst) == 0:
            answer = 1
            break
        for idx in range(0, len(s_lst)-1):
            if s_lst[idx] == s_lst[idx+1]:
                s_lst[idx], s_lst[idx+1] = 0, 0
            else:
                lst.append(s_lst[idx])
                s_lst.popleft()
        s_lst = [s for s in s_lst if s != 0]
        cnt += 1
    if answer == -1:
        answer = 0
    return answer

s = 'cdcd'
solution(s)
```
