```python
def solution(decimal):
    lst = ''
    # 정수가 0보다 크거나 같은 경우
    while decimal > 0:
        tmp = decimal % 2
        # 정수를 2로 나눈 나머지가 0일 때 - 0 추가
        if tmp == 0: lst += '0'
        # 정수를 2로 나눈 나머지가 1일 때 - 1추가
        else: lst += '1'

        # 정수 갱신
        decimal //= 2

        answer = lst[::-1]
    return answer

print(solution(10))
print(solution(27))
print(solution(12345))
```

## **문제 10: 괄호 회전하기**

https://school.programmers.co.kr/learn/courses/30/lessons/76502

6-3) 몸풀기 문제 08번 (괄호 짝 맞추기)도 그렇고 얘도 그렇고 스택... 활용하는게 왜이렇게 어려운지모르겟음,,,,ㅜ

```python
def solution(s):
    answer = 0
    idx = len(s)-1
    while idx <= len(s):
        for i in range(0, len(s), 2):
            if s[i:i+2] in ['()', '[]', '{}']:
                answer += 1
    return answer

print(solution("[](){}"))
print(solution("}]()[{"))
print(solution("[)(]"))
print(solution("}}}"))
```
