## **문제 15 요세푸스 문제**

```python
n = 5
k = 2

def solution(n, k):
    answer = 0
    arr = [i for i in range(n)]

    while len(arr) > 1:
        arr = arr[k:] + arr[:k-1]

    return arr.pop() + 1

solution(n, k)
```

## **문제 16: 기능 개발**

https://school.programmers.co.kr/learn/courses/30/lessons/42586

```python
def solution(progresses, speeds):
    import math
    answer = []
    time = []
    for i in range(len(progresses)):
        temp = math.ceil((100 - progresses[i]) / speeds[i])
        time.append(temp)
    cnt = 0
    print(time)
    for i in range(len(time)-1):
        if i == len(time) - 2:
                print(i, ': 마지막 인덱스')
                answer.append(cnt+1)
                cnt = 0
        elif time[i] >= time[i+1]:
            print('기존', time)
            time[i], time[i+1] = time[i+1], time[i]
            cnt += 1
            print(i,'번째 인덱스')
            print('변경횟수',  cnt)
            if i == len(time) - 2:
                print(i, ': 마지막 인덱스')
                answer.append(cnt+1)
                cnt = 0
        else:
            answer.append(cnt+1)
            cnt = 0
            print('답안 추가', answer)
    return answer
```

```python
progresses = [93, 30, 55]
speeds = [1, 30, 5]

print(solution(progresses, speeds))
```
